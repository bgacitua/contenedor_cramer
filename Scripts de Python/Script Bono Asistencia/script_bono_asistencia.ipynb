```python
import numpy as np
import pandas as pd
from IPython.display import display
```

```python
#IMPORTAR ARCHIVOS PARA CRUZAR LAS TABLAS
reporte_bonificacion_combinado = r"C:\Users\bgacitua\Desktop\Reportes Corrección de Marcas\Reportes_marcas_rflex\reporte_turnos_sin_marcas_20250904090441.xlsx"
reporte_turnos_bruto = r"C:\Users\bgacitua\Desktop\Reportes Corrección de Marcas\Reportes_marcas_rflex\T90S3ZYx_reporte_marca_sin_turno.xlsx"

df_principal = pd.read_csv(reporte_bonificacion_combinado)
df_auxiliar = pd.read_excel(reporte_turnos_bruto, sheet_name='Reporte')

display(df_principal.head())
display(df_auxiliar.head())
```

```python
#Filtrar turnos para el bono asistencia
puestos_df = sorted(df_principal['puesto'].dropna().unique())
#display(pd.DataFrame(puestos_df, columns=['Puesto']))

puestos_filtrados = ['Analista de Control De Calidad', 'Analista De Microbiología', 'Asistente De Laboratorio', 'Coordinadora Muestras', 'Inspector De Proceso', 'Administrativo Bodega Despacho', 'Ayudante De Bodega', 'Asistente De Bodega', 'Chofer Administrativo Transporte', 'Chofer', 'Coordinador de Planta', 'Encargado De Bodega Inflamables', 'Encargado De Bodega Materias Primas', 'Operario', 'Operario Almacenamiento y Gestión de Residuos', 'Peoneta']

df_filtrado = df_principal[df_principal['puesto'].isin(puestos_filtrados)]
print("Dataframe de puestos filtrados:")
display(df_filtrado)

```

```python
df_filtrado.columns
```

```python
#SEPARAMOS LA COLUMNA 'marca entrada' EN DOS COLUMNAS: 'fecha_entrada' Y 'hora_entrada'

df_filtrado['marca_entrada_dt'] = pd.to_datetime(
    df_filtrado['marca entrada'],
    format='%Y-%m-%d %H:%M:%S', # <-- ¡Ajusta este formato según tus datos reales!
    errors='coerce'
)

# 2. Crear la columna 'fecha_entrada'
df_filtrado['fecha_entrada'] = df_filtrado.apply(
    lambda row: "Sin registro" if pd.isna(row['marca_entrada_dt']) else str(row['marca_entrada_dt'].date()),
    axis=1
)

# 3. Crear la columna 'marca_entrada_1' (hora)
#    Si 'marca_entrada_dt' es NaT, la nueva columna tendrá "Sin registro".
#    De lo contrario, extraerá la hora y la convertirá a string.
df_filtrado['marca_entrada_1'] = df_filtrado.apply(
    lambda row: "Sin registro" if pd.isna(row['marca_entrada_dt']) else str(row['marca_entrada_dt'].time()),
    axis=1
)

# Opcional: Eliminar la columna temporal 'marca_entrada_dt' si ya no la necesitas
df_filtrado = df_filtrado.drop(columns=['marca_entrada_dt'])

# --- FIN DEL CÓDIGO CORREGIDO ---

print(df_filtrado[['marca entrada', 'fecha_entrada', 'marca_entrada_1']].head())
```

```python
    #SEPARAMOS LA COLUMNA 'marca salida' EN DOS COLUMNAS: 'fecha_salida' Y 'hora_salida_1'

df_filtrado['marca_salida_dt'] = pd.to_datetime(
    df_filtrado['marca salida'],
    format='%Y-%m-%d %H:%M:%S', # <-- ¡Ajusta este formato según tus datos reales!
    errors='coerce'
)

# 2. Crear la columna 'fecha_salida'
df_filtrado['fecha_salida'] = df_filtrado.apply(
    lambda row: "Sin registro" if pd.isna(row['marca_salida_dt']) else str(row['marca_salida_dt'].date()),
    axis=1
)

# 3. Crear la columna 'marca_salida_1' (hora)
#    Si 'marca_salida_dt' es NaT, la nueva columna tendrá "Sin registro".
#    De lo contrario, extraerá la hora y la convertirá a string.
df_filtrado['marca_salida_1'] = df_filtrado.apply(
    lambda row: "Sin registro" if pd.isna(row['marca_salida_dt']) else str(row['marca_salida_dt'].time()),
    axis=1
)

# Opcional: Eliminar la columna temporal 'marca_entrada_dt' si ya no la necesitas
df_filtrado = df_filtrado.drop(columns=['marca_salida_dt'])

# --- FIN DEL CÓDIGO CORREGIDO ---

print(df_filtrado[['marca salida', 'fecha_salida', 'marca_salida_1']].head())
```

```python
df_filtrado.columns
```

```python
# Reordenar columnas para mostrar fecha_entrada y marca_entrada_1 después de "marca entrada",
# y fecha_salida y marca_salida_1 después de "marca salida"
columnas = list(df_filtrado.columns)
for col in ['fecha_entrada', 'marca_entrada_1']:
    columnas.remove(col)
for col in ['fecha_salida', 'marca_salida_1']:
    columnas.remove(col)

# Insertar después de "marca entrada"
idx_entrada = columnas.index('marca entrada') + 1
columnas[idx_entrada:idx_entrada] = ['fecha_entrada', 'marca_entrada_1']

# Insertar después de "marca salida"
idx_salida = columnas.index('marca salida') + 1
columnas[idx_salida:idx_salida] = ['fecha_salida', 'marca_salida_1']

df_filtrado = df_filtrado[columnas]
display(df_filtrado.head())
```
