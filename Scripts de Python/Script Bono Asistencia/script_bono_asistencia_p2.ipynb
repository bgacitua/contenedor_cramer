```python
# Automatizaci√≥n de detecci√≥n de atrasos y omisiones de marca
import datetime as dt
from datetime import timedelta

# Convertir columnas de tiempo a formato datetime si no lo est√°n
df_filtrado_puestos['horaInicio'] = pd.to_datetime(df_filtrado_puestos['horaInicio'], format='%H:%M:%S', errors='coerce').dt.time
df_filtrado_puestos['horaProgramadaInicio'] = pd.to_datetime(df_filtrado_puestos['horaProgramadaInicio'], errors='coerce')
df_filtrado_puestos['HoraInicioMarcaReloj'] = pd.to_datetime(df_filtrado_puestos['HoraInicioMarcaReloj'], errors='coerce')
df_filtrado_puestos['HoraTerminoMarcaReloj'] = pd.to_datetime(df_filtrado_puestos['HoraTerminoMarcaReloj'], errors='coerce')
df_filtrado_puestos['fecha'] = pd.to_datetime(df_filtrado_puestos['fecha'], errors='coerce')

# Funci√≥n para detectar atrasos con verificaci√≥n de hora programada espec√≠fica y manejo de turnos nocturnos
def detectar_atraso(hora_inicio, hora_marca_reloj, hora_programada_inicio=None, flexibilidad_minutos=10):
    """
    Detecta si hay atraso considerando flexibilidad de 10 minutos y turnos nocturnos.
    Primero compara con horaInicio, pero si hay atraso, verifica horaProgramadaInicio
    como segunda instancia antes de confirmar el atraso.
    
    MANEJO DE TURNOS NOCTURNOS:
    - Si la hora programada es cerca de medianoche (ej: 00:00:01) y la marca es del d√≠a anterior (ej: 23:55:47),
      se considera que el trabajador lleg√≥ temprano, no tarde.
    """
    if pd.isna(hora_inicio) or pd.isna(hora_marca_reloj):
        return False
    
    fecha_marca = hora_marca_reloj.date()
    
    # Primera verificaci√≥n: comparar con horaInicio (hora est√°ndar del turno)
    hora_limite_estandar = dt.datetime.combine(fecha_marca, hora_inicio) + timedelta(minutes=flexibilidad_minutos)
    
    # MANEJO ESPECIAL PARA TURNOS NOCTURNOS con horaInicio
    # Si la hora de inicio es muy temprano (cerca de medianoche) y la marca es muy tarde (cerca de medianoche del d√≠a anterior)
    if hora_inicio.hour <= 6:  # Horario nocturno (entre 00:00 y 06:00)
        if hora_marca_reloj.hour >= 18:  # Marca del d√≠a anterior (despu√©s de las 18:00)
            # El trabajador marc√≥ el d√≠a anterior para un turno que inicia despu√©s de medianoche
            # Ajustar la hora l√≠mite al d√≠a anterior
            hora_limite_estandar = dt.datetime.combine(fecha_marca - timedelta(days=1), hora_inicio) + timedelta(minutes=flexibilidad_minutos)
            # Mover la hora l√≠mite al d√≠a siguiente (cruzar medianoche)
            hora_limite_estandar += timedelta(days=1)
    
    # Si no hay atraso con la hora est√°ndar, no es atraso
    if hora_marca_reloj <= hora_limite_estandar:
        return False
    
    # Segunda verificaci√≥n: si hay hora programada espec√≠fica para ese d√≠a, usarla
    if pd.notna(hora_programada_inicio):
        # Extraer solo la hora de horaProgramadaInicio
        if isinstance(hora_programada_inicio, pd.Timestamp):
            hora_programada_especifica = hora_programada_inicio.time()
            fecha_programada = hora_programada_inicio.date()
        else:
            hora_programada_especifica = pd.to_datetime(hora_programada_inicio).time()
            fecha_programada = pd.to_datetime(hora_programada_inicio).date()
        
        # Comparar con la hora programada espec√≠fica
        hora_limite_especifica = dt.datetime.combine(fecha_programada, hora_programada_especifica) + timedelta(minutes=flexibilidad_minutos)
        
        # MANEJO ESPECIAL PARA TURNOS NOCTURNOS con horaProgramadaInicio
        # Si la hora programada es muy temprano (cerca de medianoche) y la marca es muy tarde (cerca de medianoche del d√≠a anterior)
        if hora_programada_especifica.hour <= 6:  # Horario nocturno (entre 00:00 y 06:00)
            if hora_marca_reloj.hour >= 18:  # Marca del d√≠a anterior (despu√©s de las 18:00)
                # El trabajador marc√≥ el d√≠a anterior para un turno que inicia despu√©s de medianoche
                # No es atraso, lleg√≥ temprano
                return False
        
        # Si lleg√≥ antes del l√≠mite de la hora espec√≠fica, no es atraso
        if hora_marca_reloj <= hora_limite_especifica:
            return False
    
    # Es atraso solo si excede ambas verificaciones
    return True

# Funci√≥n para calcular minutos de atraso con verificaci√≥n de hora programada espec√≠fica y manejo de turnos nocturnos
def calcular_minutos_atraso(hora_inicio, hora_marca_reloj, hora_programada_inicio=None, flexibilidad_minutos=10):
    """
    Calcula los minutos de atraso considerando la hora programada espec√≠fica y turnos nocturnos
    """
    if pd.isna(hora_inicio) or pd.isna(hora_marca_reloj):
        return 0
    
    fecha_marca = hora_marca_reloj.date()
    
    # Determinar qu√© hora usar como referencia
    hora_referencia = hora_inicio
    fecha_referencia = fecha_marca
    
    # Si hay hora programada espec√≠fica, usarla como referencia
    if pd.notna(hora_programada_inicio):
        if isinstance(hora_programada_inicio, pd.Timestamp):
            hora_referencia = hora_programada_inicio.time()
            fecha_referencia = hora_programada_inicio.date()
        else:
            hora_referencia = pd.to_datetime(hora_programada_inicio).time()
            fecha_referencia = pd.to_datetime(hora_programada_inicio).date()
    
    hora_limite = dt.datetime.combine(fecha_referencia, hora_referencia) + timedelta(minutes=flexibilidad_minutos)
    
    # MANEJO ESPECIAL PARA TURNOS NOCTURNOS
    # Si la hora de referencia es muy temprano (cerca de medianoche) y la marca es muy tarde (cerca de medianoche del d√≠a anterior)
    if hora_referencia.hour <= 6:  # Horario nocturno (entre 00:00 y 06:00)
        if hora_marca_reloj.hour >= 18:  # Marca del d√≠a anterior (despu√©s de las 18:00)
            # El trabajador marc√≥ el d√≠a anterior para un turno que inicia despu√©s de medianoche
            # No hay atraso, lleg√≥ temprano
            return 0
    
    if hora_marca_reloj > hora_limite:
        return int((hora_marca_reloj - hora_limite).total_seconds() / 60)
    return 0

# Aplicar las funciones para detectar atrasos con verificaci√≥n de hora programada espec√≠fica
df_filtrado_puestos['atraso_entrada'] = df_filtrado_puestos.apply(
    lambda row: detectar_atraso(row['horaInicio'], row['HoraInicioMarcaReloj'], row['horaProgramadaInicio']), axis=1
)

df_filtrado_puestos['minutos_atraso_entrada'] = df_filtrado_puestos.apply(
    lambda row: calcular_minutos_atraso(row['horaInicio'], row['HoraInicioMarcaReloj'], row['horaProgramadaInicio']), axis=1
)

# Detectar omisiones de marca - pero solo cuando NO hay ausencias justificadas
# Definir ausencias justificadas que no deber√≠an generar omisiones
ausencias_justificadas = [
    'Vacaci√≥n', 'Vacaci√≥n Progresiva', 'Licencia', 'Licencia Accidente Trayecto', 'Licencia Accidente Trabajo','Licencia Ni√±o Menor', 
    'Licencia postnatal', 'Permiso con goce de sueldo', 'Permiso por Fallecimiento'
]

# Detectar omisiones solo cuando NO hay ausencias justificadas
df_filtrado_puestos['omision_marca_entrada'] = (
    df_filtrado_puestos['HoraInicioMarcaReloj'].isna() & 
    ~df_filtrado_puestos['nombreTipoAusencia'].isin(ausencias_justificadas)
)

df_filtrado_puestos['omision_marca_salida'] = (
    df_filtrado_puestos['HoraTerminoMarcaReloj'].isna() & 
    ~df_filtrado_puestos['nombreTipoAusencia'].isin(ausencias_justificadas)
)

print("Columnas creadas:")
print("- atraso_entrada: True si hay atraso (considerando 10 min de flexibilidad)")
print("  * Primera verificaci√≥n: compara con horaInicio (hora est√°ndar)")
print("  * Segunda verificaci√≥n: si hay atraso, revisa horaProgramadaInicio (hora espec√≠fica del d√≠a)")
print("- minutos_atraso_entrada: Cantidad de minutos de atraso (basado en hora m√°s apropiada)")
print("- omision_marca_entrada: True si no marc√≥ entrada Y no tiene ausencia justificada")
print("- omision_marca_salida: True si no marc√≥ salida Y no tiene ausencia justificada")
print("- Ausencias justificadas excluidas: Vacaci√≥n, Vacaci√≥n Progresiva, Licencias, Permisos")

# Mostrar estad√≠sticas
print(f"\n--- Estad√≠sticas de atrasos y omisiones ---")
print(f"Total de registros analizados: {len(df_filtrado_puestos)}")
print(f"Atrasos de entrada detectados: {df_filtrado_puestos['atraso_entrada'].sum()}")
print(f"Omisiones de marca de entrada: {df_filtrado_puestos['omision_marca_entrada'].sum()}")
print(f"Omisiones de marca de salida: {df_filtrado_puestos['omision_marca_salida'].sum()}")
print(f"Promedio de minutos de atraso (cuando hay atraso): {df_filtrado_puestos[df_filtrado_puestos['atraso_entrada']]['minutos_atraso_entrada'].mean():.1f}")

# Mostrar ejemplos
print("\n--- Ejemplos de atrasos detectados ---")
atrasos = df_filtrado_puestos[df_filtrado_puestos['atraso_entrada']].head(5)
display(atrasos[['rut', 'nombre', 'fecha', 'horaInicio', 'HoraInicioMarcaReloj', 'atraso_entrada', 'minutos_atraso_entrada']])

print("\n--- Ejemplos de omisiones de marca de entrada ---")
omisiones_entrada = df_filtrado_puestos[df_filtrado_puestos['omision_marca_entrada']].head(5)
display(omisiones_entrada[['rut', 'nombre', 'fecha', 'horaInicio', 'HoraInicioMarcaReloj', 'nombreTipoAusencia', 'omision_marca_entrada']])

# Verificaci√≥n espec√≠fica para el RUT 156647713 mencionado
print(f"\n--- Verificaci√≥n espec√≠fica para RUT 156647713 ---")
rut_especifico = df_filtrado_puestos[df_filtrado_puestos['rut'] == 156647713]
if len(rut_especifico) > 0:
    omisiones_antes = rut_especifico['HoraInicioMarcaReloj'].isna().sum()
    omisiones_despues = rut_especifico['omision_marca_entrada'].sum()
    vacaciones = rut_especifico[rut_especifico['nombreTipoAusencia'].isin(['Vacaci√≥n', 'Vacaci√≥n Progresiva'])]
    print(f"Marcas de entrada ausentes (total): {omisiones_antes}")
    print(f"Omisiones v√°lidas (sin ausencias justificadas): {omisiones_despues}")
    print(f"D√≠as con vacaciones (ambos tipos): {len(vacaciones)}")
    if len(vacaciones) > 0:
        print("Fechas con vacaciones:")
        display(vacaciones[['fecha', 'nombreTipoAusencia', 'HoraInicioMarcaReloj', 'omision_marca_entrada']])
else:
    print("RUT 156647713 no encontrado en los datos filtrados")
```

```python
# üéØ C√ÅLCULO FINAL CORRECTO DEL BONO DE ASISTENCIA
# L√≥gica FINAL: Descuento de 0.5 por per√≠odo (sin importar cu√°ntas incidencias)

import pandas as pd
from datetime import datetime

print("üéØ CALCULANDO BONO DE ASISTENCIA - L√ìGICA FINAL CORRECTA")
print("=" * 70)

# Verificar que tenemos las variables necesarias
if 'df_final_consolidado' not in globals():
    print("‚ùå Error: df_final_consolidado no est√° definido. Ejecuta primero la celda de correcci√≥n.")
else:
    # Crear una copia del DataFrame consolidado para trabajar con la l√≥gica FINAL
    df_bono_final = df_final_consolidado.copy()

    print(f"üìÖ PER√çODOS DE EVALUACI√ìN:")
    print(f"   ‚Ä¢ Per√≠odo 1: {periodo_1_inicio.strftime('%d-%m-%Y')} al {periodo_1_fin.strftime('%d-%m-%Y')} ‚Üí Descuento m√°ximo: 0.5")
    print(f"   ‚Ä¢ Per√≠odo 2: {periodo_2_inicio.strftime('%d-%m-%Y')} al {periodo_2_fin.strftime('%d-%m-%Y')} ‚Üí Descuento m√°ximo: 0.5")

    print(f"\nüîß L√ìGICA APLICADA:")
    print(f"   ‚Ä¢ Bono inicial: 1.0")
    print(f"   ‚Ä¢ Si tiene incidencias en per√≠odo 1: -0.5 (m√°ximo)")
    print(f"   ‚Ä¢ Si tiene incidencias en per√≠odo 2: -0.5 (m√°ximo)")
    print(f"   ‚Ä¢ Bono final = 1.0 - descuento_p1 - descuento_p2")

    # Funci√≥n FINAL para evaluar incidencias por per√≠odo
    def calcular_bono_final(rut_trabajador):
        """
        Calcula el bono final basado en descuentos por per√≠odo
        - Per√≠odo 1: 0.5 de descuento si tiene cualquier incidencia
        - Per√≠odo 2: 0.5 de descuento si tiene cualquier incidencia
        - Bono = 1.0 - descuento_p1 - descuento_p2
        """
        # Verificar que tenemos datos para evaluar
        if 'df_filtrado_puestos' not in globals():
            return {
                'rut': rut_trabajador,
                'bono_final': 1.0,  # Bono completo por defecto si no hay datos
                'descuento_p1': 0.0,
                'descuento_p2': 0.0,
                'incidencias_p1': [],
                'incidencias_p2': [],
                'motivo': 'Sin datos de evaluaci√≥n disponibles'
            }
        
        registros_trabajador = df_filtrado_puestos[df_filtrado_puestos['rut'] == rut_trabajador].copy()
        
        if len(registros_trabajador) == 0:
            return {
                'rut': rut_trabajador,
                'bono_final': 1.0,  # Bono completo si no hay registros de incidencias
                'descuento_p1': 0.0,
                'descuento_p2': 0.0,
                'incidencias_p1': [],
                'incidencias_p2': [],
                'motivo': 'Sin registros de incidencias'
            }
        
        registros_trabajador['fecha'] = pd.to_datetime(registros_trabajador['fecha'])
        
        # Inicializar descuentos
        descuento_p1 = 0.0
        descuento_p2 = 0.0
        
        # Listas para almacenar detalles de incidencias
        incidencias_p1 = []
        incidencias_p2 = []
        
        # EVALUACI√ìN PER√çODO 1 (06-07-2025 a 20-07-2025)
        registros_p1 = registros_trabajador[
            (registros_trabajador['fecha'] >= periodo_1_inicio) &
            (registros_trabajador['fecha'] <= periodo_1_fin)
        ]
        
        if len(registros_p1) > 0:
            # Verificar tipos de incidencias en P1
            licencias_p1 = registros_p1[registros_p1['nombreTipoAusencia'].isin(tipos_licencia)].shape[0]
            permisos_p1 = registros_p1[registros_p1['nombreTipoAusencia'].isin(tipos_permisos)].shape[0]
            otros_p1 = registros_p1[~registros_p1['nombreTipoAusencia'].isin(tipos_licencia + tipos_permisos + ausencias_justificadas)].shape[0]
            
            if licencias_p1 > 0 or permisos_p1 > 0 or otros_p1 > 0:
                descuento_p1 = 0.5
                if licencias_p1 > 0:
                    incidencias_p1.append(f"Licencias: {licencias_p1}")
                if permisos_p1 > 0:
                    incidencias_p1.append(f"Permisos: {permisos_p1}")
                if otros_p1 > 0:
                    incidencias_p1.append(f"Otras incidencias: {otros_p1}")
        
        # EVALUACI√ìN PER√çODO 2 (21-07-2025 a 05-08-2025)
        registros_p2 = registros_trabajador[
            (registros_trabajador['fecha'] >= periodo_2_inicio) &
            (registros_trabajador['fecha'] <= periodo_2_fin)
        ]
        
        if len(registros_p2) > 0:
            # Verificar tipos de incidencias en P2
            licencias_p2 = registros_p2[registros_p2['nombreTipoAusencia'].isin(tipos_licencia)].shape[0]
            permisos_p2 = registros_p2[registros_p2['nombreTipoAusencia'].isin(tipos_permisos)].shape[0]
            otros_p2 = registros_p2[~registros_p2['nombreTipoAusencia'].isin(tipos_licencia + tipos_permisos + ausencias_justificadas)].shape[0]
            
            if licencias_p2 > 0 or permisos_p2 > 0 or otros_p2 > 0:
                descuento_p2 = 0.5
                if licencias_p2 > 0:
                    incidencias_p2.append(f"Licencias: {licencias_p2}")
                if permisos_p2 > 0:
                    incidencias_p2.append(f"Permisos: {permisos_p2}")
                if otros_p2 > 0:
                    incidencias_p2.append(f"Otras incidencias: {otros_p2}")
        
        # Calcular bono final
        bono_final = 1.0 - descuento_p1 - descuento_p2
        
        # Determinar motivo
        if bono_final == 1.0:
            motivo = "Sin incidencias"
        elif bono_final == 0.5:
            if descuento_p1 > 0 and descuento_p2 == 0:
                motivo = "Incidencias solo en per√≠odo 1"
            elif descuento_p2 > 0 and descuento_p1 == 0:
                motivo = "Incidencias solo en per√≠odo 2"
            else:
                motivo = "Incidencias en per√≠odo indeterminado"
        else:
            motivo = "Incidencias en ambos per√≠odos"
        
        return {
            'rut': rut_trabajador,
            'bono_final': bono_final,
            'descuento_p1': descuento_p1,
            'descuento_p2': descuento_p2,
            'incidencias_p1': incidencias_p1,
            'incidencias_p2': incidencias_p2,
            'motivo': motivo
        }

    # Calcular bono para todos los trabajadores
    print(f"\nüîÑ CALCULANDO BONO FINAL PARA {len(df_bono_final)} TRABAJADORES...")
    resultados_finales = []

    for idx, row in df_bono_final.iterrows():
        resultado = calcular_bono_final(row['rut'])
        resultados_finales.append(resultado)
        
        # Mostrar progreso cada 10 trabajadores
        if (idx + 1) % 10 == 0 or (idx + 1) == len(df_bono_final):
            print(f"   Procesados: {idx + 1}/{len(df_bono_final)} trabajadores")

    # Agregar la columna de bono final
    df_bono_final['Bono_asistencia'] = [r['bono_final'] for r in resultados_finales]

    # Reordenar columnas para poner "Bono_asistencia" al final
    columnas_reordenadas = [col for col in df_bono_final.columns if col != 'Bono_asistencia']
    columnas_reordenadas.append('Bono_asistencia')
    df_bono_final = df_bono_final[columnas_reordenadas]

    # Crear DataFrame con resultados detallados
    df_resultados_finales = pd.DataFrame(resultados_finales)

    print(f"\nüìä DISTRIBUCI√ìN FINAL DEL BONO:")
    distribucion_final = df_bono_final['Bono_asistencia'].value_counts().sort_index(ascending=False)
    for valor, cantidad in distribucion_final.items():
        porcentaje = (cantidad / len(df_bono_final)) * 100
        descripcion = {
            1.0: "Bono completo (sin incidencias)",
            0.5: "Bono parcial (incidencias en 1 per√≠odo)",
            0.0: "Sin bono (incidencias en ambos per√≠odos)"
        }.get(valor, f"Bono {valor}")
        print(f"   ‚Ä¢ {descripcion}: {cantidad} trabajadores ({porcentaje:.1f}%)")

    print(f"\n‚úÖ C√ÅLCULO FINAL COMPLETADO")
    print(f"   üìÅ DataFrame final disponible como 'df_bono_final'")
    print(f"   üîç Resultados detallados disponibles como 'df_resultados_finales'")
    print(f"   üìä L√≥gica aplicada: Bono = 1.0 - descuento_p1 - descuento_p2")
    print(f"   üéØ Descuento m√°ximo por per√≠odo: 0.5")
```

```python
# üîÑ INTEGRACI√ìN DE DATOS DE OMISIONES AL C√ÅLCULO DEL BONO
import pandas as pd
from datetime import datetime

print("üîÑ INTEGRACI√ìN DE DATOS DE OMISIONES AL C√ÅLCULO DEL BONO")
print("=" * 60)

# Verificar que tenemos los datos cargados
if 'df_omisiones_adicional' not in globals():
    print(" ERROR: Datos de omisiones no cargados")
    print("   üîß Ejecuta primero la celda anterior para cargar el archivo")
    print("   üìÇ Aseg√∫rate de pegar la ruta correcta del archivo")
else:
    print("‚úÖ Datos de omisiones disponibles")
    print(f"   üìä Registros: {len(df_omisiones_adicional)}")
    
    # Detectar columna RUT
    columnas_rut = ['rut', 'RUT', 'Rut']
    columna_rut = None
    for col in columnas_rut:
        if col in df_omisiones_adicional.columns:
            columna_rut = col
            break
    
    if not columna_rut:
        print(" ERROR: No se encontr√≥ columna RUT en el archivo")
        print(f"   üìã Columnas disponibles: {list(df_omisiones_adicional.columns)}")
    else:
        print(f"   üÜî Columna RUT detectada: '{columna_rut}'")
        
        # Renombrar la columna RUT para consistencia
        if columna_rut != 'rut':
            df_omisiones_adicional = df_omisiones_adicional.rename(columns={columna_rut: 'rut'})
            print(f"   üîÑ Columna renombrada: '{columna_rut}' ‚Üí 'rut'")
        
        # Verificar que tenemos df_resultados_finales
        if 'df_resultados_finales' not in globals():
            print(" ERROR: df_resultados_finales no disponible")
            print("   üîß Ejecuta las celdas de c√°lculo del bono primero")
        else:
            print(f"‚úÖ DataFrame de resultados disponible: {len(df_resultados_finales)} registros")
            
            # Hacer backup de los datos originales
            df_resultados_originales = df_resultados_finales.copy()
            
            # Detectar columnas de omisiones disponibles
            columnas_omisiones = ['omisiones_entrada_p1', 'omisiones_entrada_p2', 
                                'omisiones_salida_p1', 'omisiones_salida_p2']
            
            columnas_disponibles = []
            for col in columnas_omisiones:
                if col in df_omisiones_adicional.columns:
                    columnas_disponibles.append(col)
            
            print(f"\nüìä COLUMNAS DE OMISIONES DISPONIBLES:")
            for col in columnas_disponibles:
                unique_vals = df_omisiones_adicional[col].value_counts().head(5)
                print(f"   üìà {col}: {len(df_omisiones_adicional[col].dropna())} valores v√°lidos")
                print(f"      Distribuci√≥n: {dict(unique_vals)}")
            
            if not columnas_disponibles:
                print("\n ERROR: No se encontraron columnas de omisiones v√°lidas")
                print("   üìã Columnas esperadas:", columnas_omisiones)
            else:
                # Realizar merge con los datos de omisiones
                print(f"\nüîÑ INTEGRANDO DATOS...")
                
                # Preparar DataFrame para merge
                columnas_merge = ['rut'] + columnas_disponibles
                df_omisiones_merge = df_omisiones_adicional[columnas_merge].copy()
                
                # Asegurar que RUT sea del mismo tipo
                df_omisiones_merge['rut'] = df_omisiones_merge['rut'].astype(int)
                
                print(f"   üìã Columnas a integrar: {columnas_disponibles}")
                print(f"   üë• Registros con omisiones: {len(df_omisiones_merge)}")
                
                # Hacer merge con df_resultados_finales
                df_resultados_actualizados = df_resultados_finales.merge(
                    df_omisiones_merge, 
                    on='rut', 
                    how='left',
                    suffixes=('_original', '_nuevo')
                )
                
                print(f"   ‚úÖ Merge completado: {len(df_resultados_actualizados)} registros")
                
                # Actualizar las columnas de omisiones en los detalles
                print(f"\nüîÑ ACTUALIZANDO DETALLES DE C√ÅLCULO...")
                
                registros_actualizados = 0
                for idx, row in df_resultados_actualizados.iterrows():
                    if isinstance(row.get('detalles'), dict):
                        detalles = row['detalles'].copy()
                        
                        # Actualizar con nuevos valores si est√°n disponibles
                        for col in columnas_disponibles:
                            if col in row and pd.notna(row[col]):
                                detalles[col] = int(row[col])
                                registros_actualizados += 1
                        
                        # Actualizar el registro
                        df_resultados_actualizados.at[idx, 'detalles'] = detalles
                
                print(f"   üìä Registros actualizados: {registros_actualizados}")
                
                # Recalcular el bono con los nuevos datos de omisiones
                print(f"\nüéØ RECALCULANDO BONO CON NUEVOS DATOS...")
                
                def recalcular_bono_con_omisiones(row):
                    """Recalcula el bono considerando las nuevas omisiones"""
                    rut = row['rut']
                    detalles = row.get('detalles', {})
                    
                    # Inicializar descuentos
                    descuento_p1 = 0.0
                    descuento_p2 = 0.0
                    
                    # PER√çODO 1 - Verificar todas las incidencias
                    if (detalles.get('licencias_p1', 0) >= 1 or
                        detalles.get('permisos_p1', 0) >= 1 or
                        detalles.get('inasistencias_p1', 0) >= 1 or
                        detalles.get('atrasos_p1', 0) >= 3 or
                        detalles.get('omisiones_entrada_p1', 0) >= 3 or
                        detalles.get('omisiones_salida_p1', 0) >= 3):
                        descuento_p1 = 0.5
                    
                    # PER√çODO 2 - Verificar todas las incidencias
                    if (detalles.get('licencias_p2', 0) >= 1 or
                        detalles.get('permisos_p2', 0) >= 1 or
                        detalles.get('inasistencias_p2', 0) >= 1 or
                        detalles.get('atrasos_p2', 0) >= 3 or
                        detalles.get('omisiones_entrada_p2', 0) >= 3 or
                        detalles.get('omisiones_salida_p2', 0) >= 3):
                        descuento_p2 = 0.5
                    
                    # Calcular bono final
                    bono_final = 1.0 - descuento_p1 - descuento_p2
                    
                    return {
                        'bono_final': bono_final,
                        'descuento_p1': descuento_p1,
                        'descuento_p2': descuento_p2
                    }
                
                # Aplicar rec√°lculo
                bonos_recalculados = df_resultados_actualizados.apply(recalcular_bono_con_omisiones, axis=1)
                
                # Actualizar columnas
                df_resultados_actualizados['bono_final'] = [b['bono_final'] for b in bonos_recalculados]
                df_resultados_actualizados['descuento_p1'] = [b['descuento_p1'] for b in bonos_recalculados]
                df_resultados_actualizados['descuento_p2'] = [b['descuento_p2'] for b in bonos_recalculados]
                
                # Limpiar columnas temporales del merge
                columnas_limpiar = [col for col in df_resultados_actualizados.columns 
                                  if col.endswith('_original') or col.endswith('_nuevo')]
                df_resultados_actualizados = df_resultados_actualizados.drop(columns=columnas_limpiar)
                
                # Actualizar df_resultados_finales global
                df_resultados_finales = df_resultados_actualizados.copy()
                
                print(f"‚úÖ REC√ÅLCULO COMPLETADO")
                
                # Mostrar comparaci√≥n de resultados
                print(f"\nüìä COMPARACI√ìN DE RESULTADOS:")
                
                # Distribuci√≥n original
                dist_original = df_resultados_originales['bono_final'].value_counts().sort_index(ascending=False)
                print(f"   üìà DISTRIBUCI√ìN ORIGINAL:")
                for valor, cantidad in dist_original.items():
                    print(f"      Bono {valor}: {cantidad} trabajadores")
                
                # Distribuci√≥n actualizada  
                dist_actualizada = df_resultados_finales['bono_final'].value_counts().sort_index(ascending=False)
                print(f"   üìà DISTRIBUCI√ìN ACTUALIZADA:")
                for valor, cantidad in dist_actualizada.items():
                    print(f"      Bono {valor}: {cantidad} trabajadores")
                
                # Cambios
                print(f"\nüîÑ CAMBIOS DETECTADOS:")
                cambios_detectados = 0
                for valor in [1.0, 0.5, 0.0]:
                    original = dist_original.get(valor, 0)
                    actualizado = dist_actualizada.get(valor, 0)
                    cambio = actualizado - original
                    if cambio != 0:
                        cambios_detectados += abs(cambio)
                        signo = "+" if cambio > 0 else ""
                        print(f"   üìä Bono {valor}: {signo}{cambio} trabajadores")
                
                if cambios_detectados == 0:
                    print(f"   ‚úÖ No hay cambios en la distribuci√≥n del bono")
                else:
                    print(f"   üéØ Total trabajadores afectados: {cambios_detectados // 2}")
                
                print(f"\n‚úÖ INTEGRACI√ìN COMPLETADA")
                print(f"   üìä Datos de omisiones integrados exitosamente")
                print(f"   üéØ Bono recalculado con nuevos criterios")
                print(f"   üìã df_resultados_finales actualizado")

print(f"\nüí° SIGUIENTE PASO:")
print(f"   üîÑ Regenerar el reporte completo con los datos actualizados")
print(f"   üìä Los nuevos c√°lculos incluir√°n las omisiones por per√≠odo")
```

```python
#  AN√ÅLISIS DE OMISIONES POR "OLVIDO DE MARCAJE" (SIN PERMISOS/VACACIONES/LICENCIAS)
import pandas as pd
import os
from datetime import datetime

print(" AN√ÅLISIS DE OMISIONES POR 'OLVIDO DE MARCAJE'")
print("=" * 60)
print(" SOLO SE CONSIDERAN:")
print("    Motivos = 'Olvido de Marcaje'")
print("    EXCLUYE: Permisos, Vacaciones, Licencias")

# Ruta del archivo de motivos de marca (CAMBIAR ESTA RUTA)
archivo_motivos_path = r"C:\Users\bgacitua\Desktop\Automatizaci√≥n HHEE\Reportes bono asistencia\reporte_turnos_marcas_corregidas_20250811102841.xlsx"

print(f"\n Intentando cargar archivo:")
print(f"    {archivo_motivos_path}")

# Verificar si existe el archivo
if not os.path.exists(archivo_motivos_path) or archivo_motivos_path.endswith("motivos_marca.xlsx"):
    print(f"\n  ARCHIVO NO ENCONTRADO O RUTA DE EJEMPLO")
    print(f"    INSTRUCCIONES:")
    print(f"   1. Pega la ruta completa de tu archivo Excel aqu√≠:")
    print(f"      archivo_motivos_path = r\"C:\\TU_RUTA\\archivo.xlsx\"")
    print(f"   2. Aseg√∫rate que el archivo contenga las columnas:")
    print(f"      - Rut")
    print(f"      - Fecha")
    print(f"      - Motivo marca de entrada")
    print(f"      - Motivo marca de salida")
    print(f"   3. Ejecuta esta celda nuevamente")
else:
    try:
        # Cargar archivo de motivos
        print(f"\n Cargando archivo de motivos...")
        
        # Intentar leer el archivo con diferentes hojas si es necesario
        try:
            df_motivos = pd.read_excel(archivo_motivos_path)
        except:
            # Si falla, intentar con la primera hoja espec√≠ficamente
            df_motivos = pd.read_excel(archivo_motivos_path, sheet_name=0)
        
        print(f"    Archivo cargado: {len(df_motivos)} registros")
        
        # Verificar y normalizar nombres de columnas
        columnas_disponibles = df_motivos.columns.tolist()
        print(f"\n Columnas disponibles: {columnas_disponibles}")
        
        # Mapeo de columnas (buscar variaciones)
        mapeo_columnas = {}
        
        # Buscar columna RUT
        for col in columnas_disponibles:
            col_lower = str(col).lower().strip()
            if 'rut' in col_lower:
                mapeo_columnas['rut'] = col
                break
        
        # Buscar columna Fecha
        for col in columnas_disponibles:
            col_lower = str(col).lower().strip()
            if 'fecha' in col_lower:
                mapeo_columnas['fecha'] = col
                break
        
        # Buscar columnas de motivos
        for col in columnas_disponibles:
            col_lower = str(col).lower().strip()
            if 'motivo' in col_lower and 'entrada' in col_lower:
                mapeo_columnas['motivo_entrada'] = col
            elif 'motivo' in col_lower and 'salida' in col_lower:
                mapeo_columnas['motivo_salida'] = col
        
        print(f"\n Mapeo de columnas encontradas:")
        for clave, valor in mapeo_columnas.items():
            print(f"    {clave}: '{valor}'")
        
        # Verificar columnas m√≠nimas
        if 'rut' in mapeo_columnas and 'fecha' in mapeo_columnas:
            
            # Renombrar columnas para trabajar con nombres est√°ndar
            df_motivos_clean = df_motivos.copy()
            df_motivos_clean.rename(columns={
                mapeo_columnas['rut']: 'rut',
                mapeo_columnas['fecha']: 'fecha',
                mapeo_columnas.get('motivo_entrada', 'Motivo marca de entrada'): 'motivo_entrada',
                mapeo_columnas.get('motivo_salida', 'Motivo marca de salida'): 'motivo_salida'
            }, inplace=True)
            
            # Limpiar y convertir RUT
            df_motivos_clean['rut'] = df_motivos_clean['rut'].astype(str).str.replace('.', '').str.replace('-', '').str.replace('K', '').str.replace('k', '')
            df_motivos_clean['rut'] = pd.to_numeric(df_motivos_clean['rut'], errors='coerce')
            
            # Convertir fecha
            df_motivos_clean['fecha'] = pd.to_datetime(df_motivos_clean['fecha'], errors='coerce')
            
            # Eliminar registros con datos inv√°lidos
            df_motivos_clean = df_motivos_clean.dropna(subset=['rut', 'fecha'])
            
            print(f"\n Datos limpiados: {len(df_motivos_clean)} registros v√°lidos")
            
            # FILTRAR SOLO "OLVIDO DE MARCAJE"
            print(f"\n FILTRANDO SOLO 'OLVIDO DE MARCAJE'...")
            
            # Crear m√°scara para olvido de marcaje
            mask_olvido_entrada = False
            mask_olvido_salida = False
            
            if 'motivo_entrada' in df_motivos_clean.columns:
                mask_olvido_entrada = df_motivos_clean['motivo_entrada'].str.contains('Olvido de Marcaje', case=False, na=False)
                olvidos_entrada = mask_olvido_entrada.sum()
                print(f"    Olvidos de marcaje ENTRADA: {olvidos_entrada}")
            
            if 'motivo_salida' in df_motivos_clean.columns:
                mask_olvido_salida = df_motivos_clean['motivo_salida'].str.contains('Olvido de Marcaje', case=False, na=False)
                olvidos_salida = mask_olvido_salida.sum()
                print(f"    Olvidos de marcaje SALIDA: {olvidos_salida}")
            
            # Filtrar registros con olvido de marcaje
            df_olvidos = df_motivos_clean[mask_olvido_entrada | mask_olvido_salida].copy()
            
            print(f"\n REGISTROS DE OLVIDO DE MARCAJE: {len(df_olvidos)}")
            
            if len(df_olvidos) > 0:
                
                # EXCLUIR D√çAS CON PERMISOS/VACACIONES/LICENCIAS
                print(f"\n EXCLUYENDO D√çAS CON PERMISOS/VACACIONES/LICENCIAS...")
                
                # Verificar si tenemos el DataFrame de datos principales
                if 'df_filtrado_puestos' in globals():
                    print(f"    Usando df_filtrado_puestos para verificar ausencias justificadas")
                    
                    # Tipos de ausencias justificadas a excluir
                    ausencias_justificadas = [
                        'Permiso', 'Vacaciones', 'Licencia', 'Feriado',
                        'Compensatorio', 'D√≠a Libre', 'Permiso Personal',
                        'Licencia M√©dica', 'Vacaciones Pendientes'
                    ]
                    
                    # Crear clave de cruce RUT + Fecha para df_filtrado_puestos
                    df_principales = df_filtrado_puestos.copy()
                    df_principales['fecha_dt'] = pd.to_datetime(df_principales['fecha'], errors='coerce')
                    df_principales['clave_cruce'] = df_principales['rut'].astype(str) + '_' + df_principales['fecha_dt'].dt.strftime('%Y-%m-%d')
                    
                    # Buscar registros con ausencias justificadas
                    ausencias_encontradas = pd.DataFrame()
                    for col in df_principales.columns:
                        if any(ausencia.lower() in str(col).lower() for ausencia in ausencias_justificadas):
                            registros_ausencia = df_principales[df_principales[col].notna() & (df_principales[col] != '')][['rut', 'fecha_dt', col]]
                            if len(registros_ausencia) > 0:
                                registros_ausencia['tipo_ausencia'] = col
                                registros_ausencia['clave_cruce'] = registros_ausencia['rut'].astype(str) + '_' + registros_ausencia['fecha_dt'].dt.strftime('%Y-%m-%d')
                                ausencias_encontradas = pd.concat([ausencias_encontradas, registros_ausencia[['rut', 'fecha_dt', 'tipo_ausencia', 'clave_cruce']]], ignore_index=True)
                    
                    if len(ausencias_encontradas) > 0:
                        print(f"    Ausencias justificadas encontradas: {len(ausencias_encontradas)} registros")
                        
                        # Crear clave de cruce para olvidos
                        df_olvidos['clave_cruce'] = df_olvidos['rut'].astype(str) + '_' + df_olvidos['fecha'].dt.strftime('%Y-%m-%d')
                        
                        # Excluir registros que coincidan con ausencias justificadas
                        olvidos_antes = len(df_olvidos)
                        df_olvidos_sin_ausencias = df_olvidos[~df_olvidos['clave_cruce'].isin(ausencias_encontradas['clave_cruce'])].copy()
                        olvidos_excluidos = olvidos_antes - len(df_olvidos_sin_ausencias)
                        
                        print(f"    Registros excluidos por ausencias justificadas: {olvidos_excluidos}")
                        print(f"    Olvidos v√°lidos despu√©s de exclusi√≥n: {len(df_olvidos_sin_ausencias)}")
                        
                        df_olvidos_final = df_olvidos_sin_ausencias
                    else:
                        print(f"     No se encontraron ausencias justificadas en los datos principales")
                        df_olvidos_final = df_olvidos
                else:
                    print(f"     df_filtrado_puestos no disponible - usando todos los olvidos")
                    df_olvidos_final = df_olvidos
                
                # CRUZAR CON DATOS PRINCIPALES
                print(f"\n CRUZANDO CON DATOS PRINCIPALES...")
                
                if 'df_filtrado_puestos' in globals():
                    # Preparar datos principales para cruce
                    df_principales_cruce = df_filtrado_puestos[['rut', 'nombre', 'apellidoPaterno', 'apellidoMaterno', 'PuestoIntegracion_trabajador', 'fecha']].copy()
                    df_principales_cruce['fecha'] = pd.to_datetime(df_principales_cruce['fecha'], errors='coerce')
                    df_principales_cruce = df_principales_cruce.dropna(subset=['rut', 'fecha'])
                    
                    # Hacer cruce por RUT y Fecha
                    df_cruce_olvidos = pd.merge(
                        df_olvidos_final,
                        df_principales_cruce,
                        on=['rut', 'fecha'],
                        how='inner'
                    )
                    
                    print(f"    Cruce completado: {len(df_cruce_olvidos)} registros v√°lidos")
                    
                    if len(df_cruce_olvidos) > 0:
                        
                        # CLASIFICAR POR PER√çODOS
                        periodo_1_inicio = pd.to_datetime('2025-07-06')
                        periodo_1_fin = pd.to_datetime('2025-07-20')
                        periodo_2_inicio = pd.to_datetime('2025-07-21')
                        periodo_2_fin = pd.to_datetime('2025-08-05')
                        
                        df_cruce_olvidos['periodo'] = 'Fuera de per√≠odo'
                        df_cruce_olvidos.loc[(df_cruce_olvidos['fecha'] >= periodo_1_inicio) & (df_cruce_olvidos['fecha'] <= periodo_1_fin), 'periodo'] = 'Per√≠odo 1'
                        df_cruce_olvidos.loc[(df_cruce_olvidos['fecha'] >= periodo_2_inicio) & (df_cruce_olvidos['fecha'] <= periodo_2_fin), 'periodo'] = 'Per√≠odo 2'
                        
                        # Estad√≠sticas por per√≠odo
                        registros_por_periodo = df_cruce_olvidos['periodo'].value_counts()
                        print(f"\n OLVIDOS DE MARCAJE POR PER√çODO:")
                        for periodo, cantidad in registros_por_periodo.items():
                            print(f"    {periodo}: {cantidad} registros")
                        
                        # Muestra de datos
                        print(f"\n MUESTRA DE OLVIDOS DE MARCAJE (primeros 5):")
                        for idx, row in df_cruce_olvidos.head(5).iterrows():
                            print(f"\n    RUT {int(row['rut'])} - {row['nombre']} {row['apellidoPaterno']}")
                            print(f"       Fecha: {row['fecha'].strftime('%d/%m/%Y')} - {row['periodo']}")
                            if 'motivo_entrada' in row and pd.notna(row['motivo_entrada']) and 'Olvido' in str(row['motivo_entrada']):
                                print(f"       Olvido de marcaje ENTRADA")
                            if 'motivo_salida' in row and pd.notna(row['motivo_salida']) and 'Olvido' in str(row['motivo_salida']):
                                print(f"       Olvido de marcaje SALIDA")
                        
                        print(f"\n AN√ÅLISIS DE OLVIDOS COMPLETADO")
                        print(f"    DataFrame disponible: df_cruce_olvidos")
                        print(f"    {len(df_cruce_olvidos)} olvidos de marcaje v√°lidos")
                        print(f"    Excluidos: d√≠as con permisos/vacaciones/licencias")
                        print(f"    Listo para an√°lisis por per√≠odo o exportaci√≥n")
                        
                    else:
                        print(f"\n  NO SE ENCONTRARON COINCIDENCIAS")
                        print(f"    Posibles causas:")
                        print(f"      ‚Ä¢ Fechas no coinciden entre archivos")
                        print(f"      ‚Ä¢ RUTs no coinciden")
                        print(f"      ‚Ä¢ Todos los olvidos tienen ausencias justificadas")
                        print(f"    Revisa los rangos de fechas y RUTs en ambos archivos")
                else:
                    print(f"\n ERROR: df_filtrado_puestos no disponible")
                    print(f"    Ejecuta primero las celdas de carga de datos principales")
                    
            else:
                print(f"\n  NO SE ENCONTRARON REGISTROS DE 'OLVIDO DE MARCAJE'")
                print(f"    Verifica que los motivos contengan exactamente el texto:")
                print(f"      ‚Ä¢ 'Olvido de Marcaje' (en cualquier columna de motivo)")
                
        else:
            print(f"\n ERROR: Faltan columnas m√≠nimas requeridas")
            print(f"    Se necesita al menos: RUT y Fecha")
            print(f"    Columnas encontradas: {list(mapeo_columnas.keys())}")
            
    except Exception as e:
        print(f"\n ERROR AL CARGAR ARCHIVO:")
        print(f"   {str(e)}")
        print(f"\n POSIBLES SOLUCIONES:")
        print(f"   1. Verificar que el archivo sea un Excel v√°lido (.xlsx)")
        print(f"   2. Cerrar el archivo si est√° abierto en Excel")
        print(f"   3. Verificar permisos de lectura")
        print(f"   4. Verificar que las columnas tengan los nombres exactos")

print(f"\n RECORDATORIO:")
print(f"    Pega la ruta del archivo en: archivo_motivos_path = r\"TU_RUTA_AQUI\"")
print(f"    Luego ejecuta esta celda para analizar solo olvidos de marcaje")
print(f"    Se excluir√°n autom√°ticamente d√≠as con permisos/vacaciones/licencias")
```

```python
# Procesamiento de omisiones por per√≠odo
print("üîç ANALIZANDO OMISIONES POR PER√çODO")
print("=" * 50)

def procesar_omisiones_periodo(df_omisiones, periodo, fecha_inicio, fecha_fin):
    """
    Procesa las omisiones para un per√≠odo espec√≠fico
    Solo cuenta las omisiones con motivo 'Olvido de Marcaje' para el bono
    """
    print(f"üìÖ Procesando per√≠odo {periodo}: {fecha_inicio} - {fecha_fin}")
    
    # Filtrar omisiones del per√≠odo
    omisiones_periodo = df_omisiones[
        (df_omisiones['fecha'] >= fecha_inicio) & 
        (df_omisiones['fecha'] <= fecha_fin)
    ].copy()
    
    print(f"   üìä Total omisiones encontradas: {len(omisiones_periodo)}")
    
    if len(omisiones_periodo) > 0:
        # Mostrar distribuci√≥n por motivo
        distribucion_motivos = omisiones_periodo['motivo'].value_counts()
        print(f"   üîç Distribuci√≥n por motivo:")
        for motivo, cantidad in distribucion_motivos.items():
            emoji = "‚ö†Ô∏è" if motivo == "Olvido de Marcaje" else "‚ÑπÔ∏è"
            print(f"      {emoji} {motivo}: {cantidad}")
        
        # Solo las omisiones de "Olvido de Marcaje" afectan el bono
        omisiones_bono = omisiones_periodo[omisiones_periodo['motivo'] == 'Olvido de Marcaje'].copy()
        print(f"   üéØ Omisiones que afectan bono ('Olvido de Marcaje'): {len(omisiones_bono)}")
        
        if len(omisiones_bono) > 0:
            # Contar omisiones por trabajador (solo las que afectan el bono)
            conteo_por_trabajador = omisiones_bono.groupby('rut').size().reset_index(name=f'omisiones_{periodo.lower()}')
            print(f"   üë• Trabajadores con omisiones que afectan bono: {len(conteo_por_trabajador)}")
            
            # Mostrar top trabajadores con m√°s omisiones
            if len(conteo_por_trabajador) > 0:
                top_omisiones = conteo_por_trabajador.nlargest(5, f'omisiones_{periodo.lower()}')
                print(f"   üèÜ Top trabajadores con m√°s omisiones:")
                for _, row in top_omisiones.iterrows():
                    print(f"      ‚Ä¢ RUT {row['rut']}: {row[f'omisiones_{periodo.lower()}']} omisiones")
        
        else:
            conteo_por_trabajador = pd.DataFrame(columns=['rut', f'omisiones_{periodo.lower()}'])
            print(f"   ‚úÖ No hay omisiones de 'Olvido de Marcaje' en este per√≠odo")
    else:
        conteo_por_trabajador = pd.DataFrame(columns=['rut', f'omisiones_{periodo.lower()}'])
        print(f"   ‚úÖ No hay omisiones registradas en este per√≠odo")
    
    return conteo_por_trabajador

# Procesar P1
omisiones_p1 = procesar_omisiones_periodo(df_omisiones, 'P1', fecha_inicio_p1, fecha_fin_p1)

print()  # L√≠nea en blanco para separar

# Procesar P2  
omisiones_p2 = procesar_omisiones_periodo(df_omisiones, 'P2', fecha_inicio_p2, fecha_fin_p2)

print(f"\nüìà RESUMEN DE OMISIONES:")
print(f"   üìä P1: {len(omisiones_p1)} trabajadores con omisiones que afectan bono")
print(f"   üìä P2: {len(omisiones_p2)} trabajadores con omisiones que afectan bono")
print(f"   ‚ö†Ô∏è Solo se cuentan las omisiones con motivo 'Olvido de Marcaje' para el c√°lculo del bono")
print(f"   ‚ÑπÔ∏è Las omisiones con motivo 'Otro' son informativas y no afectan el bono")
```
