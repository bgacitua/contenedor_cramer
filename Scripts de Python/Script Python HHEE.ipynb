```python
#Importar librería para manipular dataframes
import pandas as pd
```

```python
#Importar el dataframe
df_hhee = pd.read_excel(r"C:\Users\bgacitua\Desktop\Horas extras\HHEE Julio.xlsx", sheet_name="Hoja2")
display(df_hhee.head())
```

```python
# Agrupar por 'nombre' y sumar 'Horaextra50' y 'Horaextra100'
df_suma = df_hhee.groupby(['nombre', 'rut'])[['Horaextra50', 'Horaextra100']].sum().reset_index()
display(df_suma)
```

```python
df_suma = df_suma[(df_suma['Horaextra50'] + df_suma['Horaextra100']) != 0]
display(df_suma)
```

```python
import datetime
fecha_hoy = datetime.datetime.now().strftime("%d-%m-%Y_%H-%M-%S")

df_suma.to_csv(f"reporte_hhee_generado_{fecha_hoy}.csv", index=False, encoding="utf-8", sep=";")
```
