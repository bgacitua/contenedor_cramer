```python
# DIAGNÓSTICO Y CORRECCIÓN DE TIPOS DE DATOS PARA COLUMNAS HHEE
print("=== DIAGNÓSTICO DE TIPOS DE DATOS ===")

# Verificar tipos de datos actuales
columnas_numericas = ["HHEE Teoricas", "HHEE Efectivas", "Horaextra50", "Horaextra100"]

print("Tipos de datos actuales:")
for col in columnas_numericas:
    if col in df_hhee.columns:
        tipo_actual = df_hhee[col].dtype
        print(f"  {col}: {tipo_actual}")
        
        # Mostrar algunos valores para diagnosticar
        print(f"    Primeros valores: {df_hhee[col].head(3).tolist()}")
        
        # Verificar valores únicos si son pocos
        valores_unicos = df_hhee[col].nunique()
        if valores_unicos <= 10:
            print(f"    Valores únicos: {df_hhee[col].unique()}")

print(f"\n=== CORRECCIÓN DE TIPOS DE DATOS ===")

# Función para limpiar y convertir valores numéricos
def limpiar_y_convertir_numerico(serie, nombre_columna):
    print(f"Procesando columna: {nombre_columna}")
    
    # Crear copia de la serie
    serie_limpia = serie.copy()
    
    # Si ya es numérica, no hacer nada
    if pd.api.types.is_numeric_dtype(serie_limpia):
        print(f"  ✅ Ya es numérica")
        return serie_limpia
    
    # Convertir a string primero para limpiar
    serie_limpia = serie_limpia.astype(str)
    
    # Reemplazar valores problemáticos
    serie_limpia = serie_limpia.replace({
        'nan': '0',
        'NaN': '0', 
        'None': '0',
        'null': '0',
        '': '0',
        ' ': '0'
    })
    
    # Limpiar caracteres no numéricos (mantener solo números, puntos y comas)
    import re
    serie_limpia = serie_limpia.apply(lambda x: re.sub(r'[^\d.,\-]', '', str(x)) if pd.notna(x) else '0')
    
    # Reemplazar comas por puntos para formato decimal
    serie_limpia = serie_limpia.str.replace(',', '.')
    
    # Manejar valores vacíos después de la limpieza
    serie_limpia = serie_limpia.replace('', '0')
    
    try:
        # Convertir a numérico
        serie_numerica = pd.to_numeric(serie_limpia, errors='coerce')
        
        # Reemplazar NaN con 0
        serie_numerica = serie_numerica.fillna(0)
        
        print(f"  ✅ Convertida exitosamente")
        print(f"     Valores NaN encontrados y reemplazados por 0: {serie_limpia.isna().sum()}")
        print(f"     Rango: {serie_numerica.min():.2f} a {serie_numerica.max():.2f}")
        
        return serie_numerica
        
    except Exception as e:
        print(f"  ❌ Error en conversión: {e}")
        print(f"     Valores problemáticos: {serie_limpia[pd.to_numeric(serie_limpia, errors='coerce').isna()].unique()[:5]}")
        
        # Como último recurso, forzar a 0
        return pd.Series([0] * len(serie_limpia), index=serie_limpia.index)

# Aplicar corrección a cada columna numérica
for col in columnas_numericas:
    if col in df_hhee.columns:
        df_hhee[col] = limpiar_y_convertir_numerico(df_hhee[col], col)

print(f"\n=== VERIFICACIÓN POST-CORRECCIÓN ===")
print("Tipos de datos después de la corrección:")
for col in columnas_numericas:
    if col in df_hhee.columns:
        tipo_nuevo = df_hhee[col].dtype
        print(f"  {col}: {tipo_nuevo}")

# Verificar estadísticas básicas
print(f"\nEstadísticas básicas de las columnas corregidas:")
display(df_hhee[columnas_numericas].describe())
```

```python
# RESUMEN DE REGISTROS CON HHEE EFECTIVAS ENTRE 0 Y 0.50
print("=== ANÁLISIS DE HHEE EFECTIVAS ENTRE 0 Y 0.50 ===")

# Filtrar registros donde HHEE Efectivas > 0 y < 0.50
filtro_hhee_bajas = (df_hhee['HHEE Efectivas'] > 0) & (df_hhee['HHEE Efectivas'] < 0.50)

df_hhee_bajas = df_hhee[filtro_hhee_bajas].copy()

print(f"Total de registros en el DataFrame: {len(df_hhee)}")
print(f"Registros con HHEE Efectivas entre 0 y 0.50: {len(df_hhee_bajas)}")
print(f"Porcentaje: {(len(df_hhee_bajas) / len(df_hhee) * 100):.2f}%")

if len(df_hhee_bajas) > 0:
    # Definir las columnas que queremos en el resumen
    columnas_resumen = [
        'nombre', 'rut', 'unidad', 'detPuesto', 'centroCosto', 'fecha', 
        'tipo turno', 'hora inicio', 'hora termino', 'marca entrada', 
        'marca salida', 'HHEE Teoricas', 'HHEE Efectivas', 'Horaextra50', 'Horaextra100'
    ]
    
    # Verificar que todas las columnas existan
    columnas_existentes = [col for col in columnas_resumen if col in df_hhee_bajas.columns]
    columnas_faltantes = [col for col in columnas_resumen if col not in df_hhee_bajas.columns]
    
    if columnas_faltantes:
        print(f"\n⚠️  Columnas no encontradas en el DataFrame: {columnas_faltantes}")
        print(f"✅ Columnas disponibles: {columnas_existentes}")
    
    # Crear DataFrame de resumen con las columnas disponibles
    df_resumen_hhee_bajas = df_hhee_bajas[columnas_existentes].copy()
    
    # Ordenar por HHEE Efectivas (de menor a mayor)
    df_resumen_hhee_bajas = df_resumen_hhee_bajas.sort_values('HHEE Efectivas')
    
    # Estadísticas descriptivas
    print(f"\n=== ESTADÍSTICAS DE HHEE EFECTIVAS ENTRE 0 Y 0.50 ===")
    print(f"Valor mínimo: {df_resumen_hhee_bajas['HHEE Efectivas'].min():.2f}")
    print(f"Valor máximo: {df_resumen_hhee_bajas['HHEE Efectivas'].max():.2f}")
    print(f"Promedio: {df_resumen_hhee_bajas['HHEE Efectivas'].mean():.2f}")
    
    # Contar por trabajador
    if 'nombre' in columnas_existentes:
        conteo_por_trabajador = df_resumen_hhee_bajas.groupby('nombre').size().sort_values(ascending=False)
        print(f"\n=== TOP 10 TRABAJADORES CON MÁS REGISTROS DE HHEE ENTRE 0-0.50 ===")
        display(conteo_por_trabajador.head(10))
    
    # Contar por unidad/departamento
    if 'unidad' in columnas_existentes:
        conteo_por_unidad = df_resumen_hhee_bajas.groupby('unidad').size().sort_values(ascending=False)
        print(f"\n=== DISTRIBUCIÓN POR UNIDAD ===")
        display(conteo_por_unidad)
    
    # Mostrar muestra de los datos
    print(f"\n=== MUESTRA DE REGISTROS CON HHEE EFECTIVAS ENTRE 0-0.50 ===")
    display(df_resumen_hhee_bajas.head(10))
    
    # Generar archivo Excel con el resumen
    import datetime
    fecha_hoy = datetime.datetime.now().strftime("%d-%m-%Y_%H-%M-%S")
    
    nombre_archivo = f"resumen_hhee_efectivas_0-050_{fecha_hoy}.xlsx"
    ruta_completa = f"C:\\Users\\bgacitua\\Desktop\\Repositorio_GitHub\\Scripts de Python\\{nombre_archivo}"
    
    try:
        # Crear archivo Excel con múltiples hojas
        with pd.ExcelWriter(ruta_completa, engine='openpyxl') as writer:
            # Hoja principal con todos los registros
            df_resumen_hhee_bajas.to_excel(writer, sheet_name='HHEE_0-050', index=False)
            
            # Hoja con resumen por trabajador (si existe la columna nombre)
            if 'nombre' in columnas_existentes:
                resumen_trabajador = df_resumen_hhee_bajas.groupby(['nombre', 'rut']).agg({
                    'HHEE Efectivas': ['count', 'sum', 'mean'],
                    'HHEE Teoricas': 'sum',
                    'Horaextra50': 'sum',
                    'Horaextra100': 'sum'
                }).round(2)
                
                # Aplanar nombres de columnas
                resumen_trabajador.columns = ['_'.join(col).strip() for col in resumen_trabajador.columns]
                resumen_trabajador = resumen_trabajador.reset_index()
                
                resumen_trabajador.to_excel(writer, sheet_name='Resumen_por_Trabajador', index=False)
            
            # Hoja con resumen por unidad (si existe la columna)
            if 'unidad' in columnas_existentes:
                resumen_unidad = df_resumen_hhee_bajas.groupby('unidad').agg({
                    'HHEE Efectivas': ['count', 'sum', 'mean'],
                    'nombre': 'nunique'
                }).round(2)
                
                resumen_unidad.columns = ['_'.join(col).strip() for col in resumen_unidad.columns]
                resumen_unidad = resumen_unidad.reset_index()
                
                resumen_unidad.to_excel(writer, sheet_name='Resumen_por_Unidad', index=False)
        
        print(f"\n✅ ARCHIVO EXCEL GENERADO EXITOSAMENTE:")
        print(f"📁 {ruta_completa}")
        print(f"📊 {len(df_resumen_hhee_bajas)} registros exportados")
        print(f"📋 Incluye hojas: Registros principales, Resumen por trabajador, Resumen por unidad")
        
    except Exception as e:
        print(f"\n❌ Error al generar archivo Excel: {e}")
        print("💡 Intentando generar archivo CSV como alternativa...")
        
        # Alternativa: generar CSV
        nombre_csv = f"resumen_hhee_efectivas_0-050_{fecha_hoy}.csv"
        ruta_csv = f"C:\\Users\\bgacitua\\Desktop\\Repositorio_GitHub\\Scripts de Python\\{nombre_csv}"
        
        df_resumen_hhee_bajas.to_csv(ruta_csv, index=False, encoding='utf-8-sig', sep=';')
        print(f"✅ Archivo CSV generado: {ruta_csv}")

else:
    print(f"\n✅ ¡Excelente! No hay registros con HHEE Efectivas entre 0 y 0.50")
    print(f"Todos los registros tienen HHEE Efectivas = 0 o >= 0.50")

print(f"\n🎯 ANÁLISIS DE HHEE EFECTIVAS COMPLETADO")
```

```python
# APLICAR REGLAS DE HORAS EXTRAS PARA SÁBADOS
print("=== APLICANDO REGLAS DE HORAS EXTRAS PARA SÁBADOS ===")

# Primero, verificar si tenemos una columna de fecha para identificar sábados
columna_fecha = None
posibles_columnas_fecha = ['fecha', 'hora inicio', 'hora termino']

for col in posibles_columnas_fecha:
    if col in df_hhee.columns:
        # Verificar si la columna ya es datetime
        if pd.api.types.is_datetime64_any_dtype(df_hhee[col]):
            columna_fecha = col
            print(f"✅ Usando columna '{col}' para identificar sábados")
            break
        else:
            # Intentar convertir a datetime
            try:
                df_hhee[col] = pd.to_datetime(df_hhee[col], errors='coerce', dayfirst=True)
                if df_hhee[col].notna().any():
                    columna_fecha = col
                    print(f"✅ Columna '{col}' convertida a datetime para identificar sábados")
                    break
            except:
                continue

if columna_fecha is None:
    print("❌ No se encontró una columna de fecha válida para identificar sábados")
    print("Columnas disponibles:", df_hhee.columns.tolist())
else:
    # Crear columna auxiliar para identificar sábados
    df_hhee['es_sabado'] = df_hhee[columna_fecha].dt.dayofweek == 5  # 5 = Sábado
    
    # Contar registros de sábado
    registros_sabado = df_hhee['es_sabado'].sum()
    registros_con_hhee_sabado = df_hhee[df_hhee['es_sabado'] & (df_hhee['HHEE Efectivas'] > 0)].shape[0]
    
    print(f"📊 ESTADÍSTICAS DE SÁBADOS:")
    print(f"Total de registros de sábado: {registros_sabado}")
    print(f"Registros de sábado con HHEE Efectivas > 0: {registros_con_hhee_sabado}")
    
    if registros_con_hhee_sabado > 0:
        print(f"\n🔄 APLICANDO TRANSFORMACIONES...")
        
        # Contador para seguimiento
        transformaciones = 0
        
        # Filtro para registros de sábado con HHEE Efectivas
        filtro_sabado_hhee = df_hhee['es_sabado'] & (df_hhee['HHEE Efectivas'] > 0)
        
        # Mostrar muestra ANTES de las transformaciones
        print(f"\n=== MUESTRA ANTES DE TRANSFORMACIONES ===")
        muestra_antes = df_hhee[filtro_sabado_hhee][['nombre', 'fecha', 'HHEE Efectivas', 'Horaextra50', 'Horaextra100', 'es_sabado']].head(10)
        display(muestra_antes)
        
        # Para cada registro de sábado con HHEE Efectivas
        for idx in df_hhee[filtro_sabado_hhee].index:
            # Obtener valores actuales
            hhee_efectivas = df_hhee.loc[idx, 'HHEE Efectivas']
            horaextra50_actual = df_hhee.loc[idx, 'Horaextra50']
            horaextra100_actual = df_hhee.loc[idx, 'Horaextra100']
            
            # Nueva lógica:
            # 1. Todo lo que está en Horaextra50 pasa a Horaextra100
            # 2. Las HHEE Efectivas también van a Horaextra100
            
            nuevo_horaextra100 = horaextra100_actual + horaextra50_actual + hhee_efectivas
            
            # Aplicar cambios
            df_hhee.loc[idx, 'Horaextra100'] = nuevo_horaextra100
            df_hhee.loc[idx, 'Horaextra50'] = 0  # Se vacía porque todo pasa al 100%
            
            transformaciones += 1
        
        print(f"✅ TRANSFORMACIONES COMPLETADAS: {transformaciones} registros procesados")
        
        # Mostrar muestra DESPUÉS de las transformaciones
        print(f"\n=== MUESTRA DESPUÉS DE TRANSFORMACIONES ===")
        muestra_despues = df_hhee[filtro_sabado_hhee][['nombre', 'fecha', 'HHEE Efectivas', 'Horaextra50', 'Horaextra100', 'es_sabado']].head(10)
        display(muestra_despues)
        
        # Estadísticas de verificación
        print(f"\n=== VERIFICACIÓN DE RESULTADOS ===")
        total_horaextra50_sabados = df_hhee[df_hhee['es_sabado']]['Horaextra50'].sum()
        total_horaextra100_sabados = df_hhee[df_hhee['es_sabado']]['Horaextra100'].sum()
        
        print(f"Total Horaextra50 en sábados (debe ser 0): {total_horaextra50_sabados}")
        print(f"Total Horaextra100 en sábados: {total_horaextra100_sabados}")
        
        # Verificar que no quedan Horaextra50 en sábados
        sabados_con_horaextra50 = df_hhee[df_hhee['es_sabado'] & (df_hhee['Horaextra50'] > 0)].shape[0]
        
        if sabados_con_horaextra50 == 0:
            print("✅ VERIFICACIÓN EXITOSA: No quedan Horaextra50 en sábados")
        else:
            print(f"⚠️  ADVERTENCIA: Aún quedan {sabados_con_horaextra50} registros con Horaextra50 en sábados")
        
        # Resumen por trabajador en sábados
        if 'nombre' in df_hhee.columns:
            print(f"\n=== RESUMEN POR TRABAJADOR EN SÁBADOS (TOP 10) ===")
            resumen_trabajador_sabado = df_hhee[df_hhee['es_sabado']].groupby('nombre').agg({
                'HHEE Efectivas': 'sum',
                'Horaextra50': 'sum',
                'Horaextra100': 'sum'
            }).round(2)
            
            # Ordenar por Horaextra100 descendente
            resumen_trabajador_sabado = resumen_trabajador_sabado.sort_values('Horaextra100', ascending=False)
            display(resumen_trabajador_sabado.head(10))
    
    else:
        print("ℹ️  No hay registros de sábado con HHEE Efectivas para procesar")
    
    # Eliminar columna auxiliar
    df_hhee = df_hhee.drop('es_sabado', axis=1)
    print(f"\n🧹 Columna auxiliar 'es_sabado' eliminada")

print(f"\n🎯 PROCESAMIENTO DE SÁBADOS COMPLETADO")

# Mostrar estadísticas finales generales
print(f"\n=== ESTADÍSTICAS FINALES DEL DATAFRAME ===")
print(f"Total de registros: {len(df_hhee)}")
print(f"Total HHEE Efectivas: {df_hhee['HHEE Efectivas'].sum():.2f}")
print(f"Total Horaextra50: {df_hhee['Horaextra50'].sum():.2f}")
print(f"Total Horaextra100: {df_hhee['Horaextra100'].sum():.2f}")
```

```python
# GENERAR REPORTE EXCEL CON ESTRUCTURA ESPECÍFICA (USANDO DATOS ORIGINALES)
print("=== GENERANDO REPORTE EXCEL CON ESTRUCTURA ESPECÍFICA ===")

# Verificar si df_hhee_original existe, si no, cargarlo
if 'df_hhee_original' not in globals():
    print("📥 Cargando datos originales...")
    df_hhee_original = pd.read_csv(r"C:\Users\bgacitua\Desktop\Automatización HHEE\Reporte bonificacion\2025-08-19_10_41_59-reporte_combinado_detallado.csv", delimiter=',')
    
    # Aplicar corrección de tipos de datos
    columnas_numericas = ["HHEE Teoricas", "HHEE Efectivas", "Horaextra50", "Horaextra100"]
    
    # Función para limpiar y convertir valores numéricos
    def limpiar_y_convertir_numerico_simple(serie):
        if pd.api.types.is_numeric_dtype(serie):
            return serie
        
        serie_limpia = serie.astype(str)
        serie_limpia = serie_limpia.replace({
            'nan': '0', 'NaN': '0', 'None': '0', 'null': '0', '': '0', ' ': '0'
        })
        
        import re
        serie_limpia = serie_limpia.apply(lambda x: re.sub(r'[^\d.,\-]', '', str(x)) if pd.notna(x) else '0')
        serie_limpia = serie_limpia.str.replace(',', '.')
        serie_limpia = serie_limpia.replace('', '0')
        
        serie_numerica = pd.to_numeric(serie_limpia, errors='coerce')
        return serie_numerica.fillna(0)
    
    # Aplicar corrección a cada columna numérica
    for col in columnas_numericas:
        if col in df_hhee_original.columns:
            df_hhee_original[col] = limpiar_y_convertir_numerico_simple(df_hhee_original[col])
    
    print(f"✅ Datos originales cargados: {len(df_hhee_original)} registros")

# Crear timestamp para nombre de archivo
import datetime
fecha_hoy = datetime.datetime.now().strftime("%d-%m-%Y_%H-%M-%S")

# Generar archivo Excel en la carpeta "Reportes generados"
nombre_archivo_excel = f"reporte_hhee_estructurado_{fecha_hoy}.xlsx"
ruta_excel = f"C:\\Users\\bgacitua\\Desktop\\Repositorio_GitHub\\Scripts de Python\\Reportes generados\\{nombre_archivo_excel}"

print(f"Preparando datos para el archivo Excel...")

# USAR DATOS ORIGINALES (antes de eliminar registros)
# HOJA 1: DataFrame original completo
df_original = df_hhee_original.copy()
print(f"Hoja 1 - Datos originales: {len(df_original)} registros")

# HOJA 2: Filtrar HHEE Efectivas >= 0.5 (MAYOR O IGUAL que 0.5)
df_hhee_altas = df_hhee_original[df_hhee_original['HHEE Efectivas'] >= 0.5].copy()
print(f"Hoja 2 - HHEE Efectivas >= 0.5: {len(df_hhee_altas)} registros")

# HOJA 3: Filtrar HHEE Efectivas > 0 Y < 0.5 (MAYOR A 0 PERO MENOR A 0.5)
df_hhee_bajas = df_hhee_original[(df_hhee_original['HHEE Efectivas'] > 0) & (df_hhee_original['HHEE Efectivas'] < 0.5)].copy()
print(f"Hoja 3 - HHEE Efectivas > 0 y < 0.5: {len(df_hhee_bajas)} registros")

# HOJA 4: Filtrar HHEE Teóricas > 0 los días SÁBADO
# Primero, convertir la columna fecha a datetime si no lo está
if 'fecha' in df_hhee_original.columns:
    df_hhee_original['fecha'] = pd.to_datetime(df_hhee_original['fecha'], errors='coerce', dayfirst=True)
    # Filtrar: HHEE Teóricas > 0 Y día de la semana = 5 (sábado)
    df_hhee_sabados = df_hhee_original[
        (df_hhee_original['HHEE Teoricas'] > 0) & 
        (df_hhee_original['fecha'].dt.dayofweek == 5)
    ].copy()
    print(f"Hoja 4 - HHEE Teóricas > 0 en Sábados: {len(df_hhee_sabados)} registros")
    
    # APLICAR LÓGICA ESPECIAL PARA HOJA 4: RECALCULAR HORAEXTRA50 Y HORAEXTRA100
    if len(df_hhee_sabados) > 0 and 'marca entrada' in df_hhee_sabados.columns and 'marca salida' in df_hhee_sabados.columns:
        print(f"\n🔄 APLICANDO LÓGICA ESPECIAL DE HORAS EXTRAS PARA SÁBADOS...")
        
        # Convertir marcas a datetime
        df_hhee_sabados['marca entrada'] = pd.to_datetime(df_hhee_sabados['marca entrada'], errors='coerce')
        df_hhee_sabados['marca salida'] = pd.to_datetime(df_hhee_sabados['marca salida'], errors='coerce')
        
        # Definir hora límite: 10:00:00
        hora_limite = pd.Timestamp('1900-01-01 10:00:00').time()
        
        registros_procesados = 0
        
        for idx in df_hhee_sabados.index:
            marca_entrada = df_hhee_sabados.loc[idx, 'marca entrada']
            marca_salida = df_hhee_sabados.loc[idx, 'marca salida']
            hhee_teoricas = df_hhee_sabados.loc[idx, 'HHEE Teoricas']
            
            # Verificar que las marcas sean válidas
            if pd.notna(marca_entrada) and pd.notna(marca_salida):
                # Extraer solo la hora de las marcas
                hora_entrada = marca_entrada.time()
                hora_salida = marca_salida.time()
                
                # Si la marca de salida es superior a las 10:00:00
                if hora_salida > hora_limite:
                    # Convertir a datetime para calcular diferencias
                    fecha_base = pd.Timestamp('1900-01-01')
                    dt_entrada = pd.Timestamp.combine(fecha_base.date(), hora_entrada)
                    dt_salida = pd.Timestamp.combine(fecha_base.date(), hora_salida)
                    dt_limite = pd.Timestamp.combine(fecha_base.date(), hora_limite)
                    
                    # Calcular horas
                    # Horaextra50: desde entrada hasta las 10:00
                    horaextra50 = (dt_limite - dt_entrada).total_seconds() / 3600
                    # Horaextra100: desde las 10:00 hasta salida
                    horaextra100 = (dt_salida - dt_limite).total_seconds() / 3600
                    
                    # Asegurar valores no negativos
                    horaextra50 = max(0, horaextra50)
                    horaextra100 = max(0, horaextra100)
                    
                    # VALIDACIÓN CRÍTICA: La suma no puede exceder HHEE Teoricas
                    suma_total = horaextra50 + horaextra100
                    if suma_total > hhee_teoricas:
                        # Ajustar proporcionalmente para que no exceda HHEE Teoricas
                        factor_ajuste = hhee_teoricas / suma_total
                        horaextra50 = horaextra50 * factor_ajuste
                        horaextra100 = horaextra100 * factor_ajuste
                        
                        print(f"⚠️  Ajuste aplicado a {df_hhee_sabados.loc[idx, 'nombre']}: Suma original {suma_total:.2f} > HHEE Teoricas {hhee_teoricas:.2f}")
                        print(f"   Nuevo HE50: {horaextra50:.2f}, Nuevo HE100: {horaextra100:.2f}")
                    
                    # Actualizar las columnas
                    df_hhee_sabados.loc[idx, 'Horaextra50'] = round(horaextra50, 2)
                    df_hhee_sabados.loc[idx, 'Horaextra100'] = round(horaextra100, 2)
                    
                    registros_procesados += 1
                else:
                    # Si no supera las 10:00, todas las horas van a HE50, pero sin exceder HHEE Teoricas
                    fecha_base = pd.Timestamp('1900-01-01')
                    dt_entrada = pd.Timestamp.combine(fecha_base.date(), hora_entrada)
                    dt_salida = pd.Timestamp.combine(fecha_base.date(), hora_salida)
                    
                    # Calcular total de horas trabajadas
                    horas_trabajadas = (dt_salida - dt_entrada).total_seconds() / 3600
                    horas_trabajadas = max(0, horas_trabajadas)
                    
                    # Asignar todo a HE50, pero sin exceder HHEE Teoricas
                    horaextra50 = min(horas_trabajadas, hhee_teoricas)
                    horaextra100 = 0
                    
                    df_hhee_sabados.loc[idx, 'Horaextra50'] = round(horaextra50, 2)
                    df_hhee_sabados.loc[idx, 'Horaextra100'] = round(horaextra100, 2)
                    
                    registros_procesados += 1
        
        print(f"✅ Procesados {registros_procesados} registros con lógica especial de sábados")
        
        # VALIDACIÓN CRÍTICA FINAL: Verificar que ningún registro exceda HHEE Teoricas
        print(f"\n🔍 VALIDACIÓN FINAL DE CONSISTENCIA:")
        df_hhee_sabados['Suma_HE'] = df_hhee_sabados['Horaextra50'] + df_hhee_sabados['Horaextra100']
        registros_inconsistentes = df_hhee_sabados[df_hhee_sabados['Suma_HE'] > df_hhee_sabados['HHEE Teoricas']]
        
        if len(registros_inconsistentes) > 0:
            print(f"⚠️  ADVERTENCIA: {len(registros_inconsistentes)} registros aún exceden HHEE Teoricas:")
            display(registros_inconsistentes[['nombre', 'HHEE Teoricas', 'Horaextra50', 'Horaextra100', 'Suma_HE']])
        else:
            print(f"✅ VALIDACIÓN EXITOSA: Todos los registros cumplen HE50 + HE100 ≤ HHEE Teoricas")
        
        # Remover columna auxiliar
        df_hhee_sabados = df_hhee_sabados.drop('Suma_HE', axis=1)
        
        # Mostrar estadísticas después del procesamiento
        total_he50_sabados = df_hhee_sabados['Horaextra50'].sum()
        total_he100_sabados = df_hhee_sabados['Horaextra100'].sum()
        total_hhee_teoricas = df_hhee_sabados['HHEE Teoricas'].sum()
        
        print(f"📊 Nuevos totales en Hoja 4:")
        print(f"   HHEE Teoricas: {total_hhee_teoricas:.2f}")
        print(f"   Horaextra50: {total_he50_sabados:.2f}")
        print(f"   Horaextra100: {total_he100_sabados:.2f}")
        print(f"   Suma HE50+HE100: {total_he50_sabados + total_he100_sabados:.2f}")
        print(f"   ✅ Cumple regla (≤ HHEE Teoricas): {total_he50_sabados + total_he100_sabados <= total_hhee_teoricas}")
        
else:
    # Si no hay columna fecha, usar una alternativa como 'hora inicio'
    columna_fecha_alt = None
    for col in ['hora inicio', 'hora termino']:
        if col in df_hhee_original.columns:
            columna_fecha_alt = col
            break
    
    if columna_fecha_alt:
        df_hhee_original[columna_fecha_alt] = pd.to_datetime(df_hhee_original[columna_fecha_alt], errors='coerce', dayfirst=True)
        df_hhee_sabados = df_hhee_original[
            (df_hhee_original['HHEE Teoricas'] > 0) & 
            (df_hhee_original[columna_fecha_alt].dt.dayofweek == 5)
        ].copy()
        print(f"Hoja 4 - HHEE Teóricas > 0 en Sábados (usando {columna_fecha_alt}): {len(df_hhee_sabados)} registros")
    else:
        # Si no hay columnas de fecha, crear DataFrame vacío
        df_hhee_sabados = df_hhee_original.iloc[0:0].copy()  # DataFrame vacío con mismas columnas
        print(f"Hoja 4 - ADVERTENCIA: No se encontró columna de fecha, hoja estará vacía")

# Verificación de totales (ahora tendremos 3 categorías: =0, >=0.5, entre 0-0.5)
registros_cero = len(df_hhee_original[df_hhee_original['HHEE Efectivas'] == 0])
total_verificacion = len(df_hhee_altas) + len(df_hhee_bajas) + registros_cero
print(f"\nVerificación: {len(df_original)} = {len(df_hhee_altas)} (≥0.5) + {len(df_hhee_bajas)} (0<HHEE<0.5) + {registros_cero} (=0) = {total_verificacion}")

if total_verificacion == len(df_original):
    print("✅ Verificación exitosa: Los totales coinciden")
else:
    print("⚠️  Advertencia: Los totales no coinciden")

try:
    # Crear archivo Excel con las 4 hojas específicas
    with pd.ExcelWriter(ruta_excel, engine='openpyxl') as writer:
        
        # HOJA 1: Datos originales completos
        df_original.to_excel(writer, sheet_name='Datos_Originales', index=False)
        print("✅ Hoja 1 creada: 'Datos_Originales'")
        
        # HOJA 2: HHEE Efectivas >= 0.5
        df_hhee_altas.to_excel(writer, sheet_name='HHEE_Mayor_Igual_05', index=False)
        print("✅ Hoja 2 creada: 'HHEE_Mayor_Igual_05'")
        
        # HOJA 3: HHEE Efectivas > 0 y < 0.5 
        df_hhee_bajas.to_excel(writer, sheet_name='HHEE_Entre_0_y_05', index=False)
        print("✅ Hoja 3 creada: 'HHEE_Entre_0_y_05'")
        
        # HOJA 4: HHEE Teóricas > 0 en Sábados
        df_hhee_sabados.to_excel(writer, sheet_name='HHEE_Teoricas_Sabados', index=False)
        print("✅ Hoja 4 creada: 'HHEE_Teoricas_Sabados'")
    
    print(f"\n✅ ARCHIVO EXCEL GENERADO EXITOSAMENTE:")
    print(f"📁 {ruta_excel}")
    print(f"\n📊 ESTRUCTURA DEL ARCHIVO:")
    print(f"   📋 Hoja 1 - 'Datos_Originales': {len(df_original)} registros (datos completos)")
    print(f"   📋 Hoja 2 - 'HHEE_Mayor_Igual_05': {len(df_hhee_altas)} registros (HHEE ≥ 0.5)")
    print(f"   📋 Hoja 3 - 'HHEE_Entre_0_y_05': {len(df_hhee_bajas)} registros (0 < HHEE < 0.5)")
    print(f"   📋 Hoja 4 - 'HHEE_Teoricas_Sabados': {len(df_hhee_sabados)} registros (HHEE Teóricas > 0 en Sábados)")
    
    # Mostrar estadísticas por hoja
    print(f"\n=== ESTADÍSTICAS POR HOJA ===")
    
    print(f"\n📊 HOJA 1 - DATOS ORIGINALES:")
    print(f"   Total HHEE Efectivas: {df_original['HHEE Efectivas'].sum():.2f}")
    print(f"   Total Horaextra50: {df_original['Horaextra50'].sum():.2f}")
    print(f"   Total Horaextra100: {df_original['Horaextra100'].sum():.2f}")
    
    print(f"\n📊 HOJA 2 - HHEE EFECTIVAS ≥ 0.5:")
    print(f"   Total HHEE Efectivas: {df_hhee_altas['HHEE Efectivas'].sum():.2f}")
    print(f"   Total Horaextra50: {df_hhee_altas['Horaextra50'].sum():.2f}")
    print(f"   Total Horaextra100: {df_hhee_altas['Horaextra100'].sum():.2f}")
    print(f"   Trabajadores únicos: {df_hhee_altas['nombre'].nunique() if 'nombre' in df_hhee_altas.columns else 'N/A'}")
    
    print(f"\n📊 HOJA 3 - HHEE EFECTIVAS > 0 Y < 0.5:")
    print(f"   Total HHEE Efectivas: {df_hhee_bajas['HHEE Efectivas'].sum():.2f}")
    print(f"   Total Horaextra50: {df_hhee_bajas['Horaextra50'].sum():.2f}")
    print(f"   Total Horaextra100: {df_hhee_bajas['Horaextra100'].sum():.2f}")
    print(f"   Trabajadores únicos: {df_hhee_bajas['nombre'].nunique() if 'nombre' in df_hhee_bajas.columns else 'N/A'}")
    
    print(f"\n📊 HOJA 4 - HHEE TEÓRICAS > 0 EN SÁBADOS (CON LÓGICA ESPECIAL):")
    print(f"   Total HHEE Teóricas: {df_hhee_sabados['HHEE Teoricas'].sum():.2f}")
    print(f"   Total HHEE Efectivas: {df_hhee_sabados['HHEE Efectivas'].sum():.2f}")
    print(f"   Total Horaextra50 (recalculado): {df_hhee_sabados['Horaextra50'].sum():.2f}")
    print(f"   Total Horaextra100 (recalculado): {df_hhee_sabados['Horaextra100'].sum():.2f}")
    print(f"   Trabajadores únicos: {df_hhee_sabados['nombre'].nunique() if 'nombre' in df_hhee_sabados.columns else 'N/A'}")
    if len(df_hhee_sabados) > 0:
        fecha_min = df_hhee_sabados['fecha'].min() if 'fecha' in df_hhee_sabados.columns else 'N/A'
        fecha_max = df_hhee_sabados['fecha'].max() if 'fecha' in df_hhee_sabados.columns else 'N/A'
        print(f"   Rango de fechas: {fecha_min} a {fecha_max}")
        print(f"   ⚠️  NOTA: Horaextra50/100 recalculados según marcas de entrada/salida vs 10:00:00")
    else:
        print(f"   Rango de fechas: No hay datos")
    
    # Mostrar muestras de cada hoja
    print(f"\n=== MUESTRA DE CADA HOJA ===")
    
    print(f"\n📋 MUESTRA HOJA 2 (HHEE ≥ 0.5):")
    if len(df_hhee_altas) > 0:
        columnas_muestra = ['nombre', 'rut', 'fecha', 'HHEE Efectivas', 'Horaextra50', 'Horaextra100']
        columnas_disponibles = [col for col in columnas_muestra if col in df_hhee_altas.columns]
        display(df_hhee_altas[columnas_disponibles].head(5))
    else:
        print("   No hay registros en esta categoría")
    
    print(f"\n📋 MUESTRA HOJA 3 (0 < HHEE < 0.5):")
    if len(df_hhee_bajas) > 0:
        columnas_muestra = ['nombre', 'rut', 'fecha', 'HHEE Efectivas', 'Horaextra50', 'Horaextra100']
        columnas_disponibles = [col for col in columnas_muestra if col in df_hhee_bajas.columns]
        display(df_hhee_bajas[columnas_disponibles].head(5))
    else:
        print("   No hay registros en esta categoría")
    
    print(f"\n📋 MUESTRA HOJA 4 (HHEE TEÓRICAS > 0 EN SÁBADOS):")
    if len(df_hhee_sabados) > 0:
        columnas_muestra_sabados = ['nombre', 'rut', 'fecha', 'marca entrada', 'marca salida', 'HHEE Teoricas', 'HHEE Efectivas', 'Horaextra50', 'Horaextra100']
        columnas_disponibles_sabados = [col for col in columnas_muestra_sabados if col in df_hhee_sabados.columns]
        display(df_hhee_sabados[columnas_disponibles_sabados].head(5))
        
        # Mostrar también un resumen por trabajador en sábados
        if 'nombre' in df_hhee_sabados.columns:
            print(f"\n📊 RESUMEN DE TRABAJADORES CON HHEE TEÓRICAS EN SÁBADOS (CON RECÁLCULO):")
            resumen_sabados = df_hhee_sabados.groupby('nombre').agg({
                'fecha': 'count',  # Días trabajados
                'HHEE Teoricas': 'sum',
                'HHEE Efectivas': 'sum',
                'Horaextra50': 'sum',
                'Horaextra100': 'sum'
            }).round(2)
            resumen_sabados.columns = ['Días_Sábado', 'Total_HHEE_Teoricas', 'Total_HHEE_Efectivas', 'Total_HE50_Recalc', 'Total_HE100_Recalc']
            resumen_sabados = resumen_sabados.sort_values('Total_HHEE_Teoricas', ascending=False)
            display(resumen_sabados.head(10))
        
        # Mostrar algunos ejemplos de cálculos realizados
        if 'marca entrada' in df_hhee_sabados.columns and 'marca salida' in df_hhee_sabados.columns:
            print(f"\n💡 EJEMPLOS DE CÁLCULOS REALIZADOS:")
            registros_con_marcas = df_hhee_sabados.dropna(subset=['marca entrada', 'marca salida'])
            if len(registros_con_marcas) > 0:
                for i, (idx, row) in enumerate(registros_con_marcas.head(3).iterrows()):
                    entrada = row['marca entrada'].time() if pd.notna(row['marca entrada']) else 'N/A'
                    salida = row['marca salida'].time() if pd.notna(row['marca salida']) else 'N/A'
                    he50 = row['Horaextra50']
                    he100 = row['Horaextra100']
                    print(f"   Ejemplo {i+1}: {row['nombre']}")
                    print(f"     Entrada: {entrada}, Salida: {salida}")
                    print(f"     HE50: {he50}, HE100: {he100}")
    else:
        print("   No hay registros en esta categoría")
        
except Exception as e:
    print(f"\n❌ Error al generar archivo Excel: {e}")
    print("💡 Generando archivos CSV como alternativa...")
    
    # Alternativa: generar CSVs separados
    try:
        ruta_base = f"C:\\Users\\bgacitua\\Desktop\\Repositorio_GitHub\\Scripts de Python\\Reportes generados\\reporte_hhee_estructurado_{fecha_hoy}"
        
        df_original.to_csv(f"{ruta_base}_datos_originales.csv", index=False, encoding="utf-8-sig", sep=";")
        df_hhee_altas.to_csv(f"{ruta_base}_hhee_mayor_igual_05.csv", index=False, encoding="utf-8-sig", sep=";")
        df_hhee_bajas.to_csv(f"{ruta_base}_hhee_entre_0_y_05.csv", index=False, encoding="utf-8-sig", sep=";")
        df_hhee_sabados.to_csv(f"{ruta_base}_hhee_teoricas_sabados.csv", index=False, encoding="utf-8-sig", sep=";")
        
        print(f"✅ Archivos CSV generados:")
        print(f"   📄 {ruta_base}_datos_originales.csv")
        print(f"   📄 {ruta_base}_hhee_mayor_igual_05.csv")
        print(f"   📄 {ruta_base}_hhee_entre_0_y_05.csv")
        print(f"   📄 {ruta_base}_hhee_teoricas_sabados.csv")
        
    except Exception as csv_error:
        print(f"❌ Error al generar CSVs: {csv_error}")

print(f"\n🎯 REPORTE ESTRUCTURADO COMPLETADO")
print(f"📋 Archivo listo para análisis posteriores con la estructura solicitada")
```

```python
# CARGAR DATOS ORIGINALES PARA ENCONTRAR LOS REGISTROS ENTRE 0 Y 0.5
print("=== CARGANDO DATOS ORIGINALES ===")

# Cargar el archivo original nuevamente
df_hhee_original = pd.read_csv(r"C:\Users\bgacitua\Desktop\Automatización HHEE\Reporte bonificacion\2025-08-19_10_41_59-reporte_combinado_detallado.csv", delimiter=',')

print(f"Registros en archivo original: {len(df_hhee_original)}")

# Aplicar la misma corrección de tipos de datos
columnas_numericas = ["HHEE Teoricas", "HHEE Efectivas", "Horaextra50", "Horaextra100"]

# Función para limpiar y convertir valores numéricos (reutilizando)
def limpiar_y_convertir_numerico_simple(serie):
    # Si ya es numérica, no hacer nada
    if pd.api.types.is_numeric_dtype(serie):
        return serie
    
    # Convertir a string primero para limpiar
    serie_limpia = serie.astype(str)
    
    # Reemplazar valores problemáticos
    serie_limpia = serie_limpia.replace({
        'nan': '0', 'NaN': '0', 'None': '0', 'null': '0', '': '0', ' ': '0'
    })
    
    # Limpiar caracteres no numéricos y reemplazar comas por puntos
    import re
    serie_limpia = serie_limpia.apply(lambda x: re.sub(r'[^\d.,\-]', '', str(x)) if pd.notna(x) else '0')
    serie_limpia = serie_limpia.str.replace(',', '.')
    serie_limpia = serie_limpia.replace('', '0')
    
    # Convertir a numérico
    serie_numerica = pd.to_numeric(serie_limpia, errors='coerce')
    return serie_numerica.fillna(0)

# Aplicar corrección a cada columna numérica
for col in columnas_numericas:
    if col in df_hhee_original.columns:
        df_hhee_original[col] = limpiar_y_convertir_numerico_simple(df_hhee_original[col])

print(f"\n=== ANÁLISIS DE DATOS ORIGINALES ===")

# Ahora buscar registros entre 0 y 0.5 en los datos originales
registros_entre_0_y_05 = df_hhee_original[(df_hhee_original['HHEE Efectivas'] > 0) & (df_hhee_original['HHEE Efectivas'] < 0.5)]

print(f"Registros con HHEE Efectivas > 0 y < 0.5 en datos originales: {len(registros_entre_0_y_05)}")

if len(registros_entre_0_y_05) > 0:
    print(f"\n¡ENCONTRADOS! Los registros son:")
    columnas_mostrar = ['nombre', 'rut', 'fecha', 'HHEE Efectivas', 'Horaextra50', 'Horaextra100']
    columnas_disponibles = [col for col in columnas_mostrar if col in registros_entre_0_y_05.columns]
    display(registros_entre_0_y_05[columnas_disponibles])
    
    # Mostrar valores únicos
    print(f"\nValores únicos de HHEE Efectivas entre 0 y 0.5:")
    valores_unicos_entre = registros_entre_0_y_05['HHEE Efectivas'].unique()
    print(sorted(valores_unicos_entre))
    
else:
    print("No se encontraron registros entre 0 y 0.5 en los datos originales")

# Estadísticas de los datos originales
print(f"\n=== ESTADÍSTICAS DE DATOS ORIGINALES ===")
print(f"Registros con HHEE Efectivas = 0: {(df_hhee_original['HHEE Efectivas'] == 0).sum()}")
print(f"Registros con HHEE Efectivas > 0 y < 0.5: {((df_hhee_original['HHEE Efectivas'] > 0) & (df_hhee_original['HHEE Efectivas'] < 0.5)).sum()}")
print(f"Registros con HHEE Efectivas = 0.5: {(df_hhee_original['HHEE Efectivas'] == 0.5).sum()}")
print(f"Registros con HHEE Efectivas >= 0.5: {(df_hhee_original['HHEE Efectivas'] >= 0.5).sum()}")
print(f"Total: {len(df_hhee_original)}")
```
